```python
# AI Dataset Quality Analysis

This notebook explores dataset quality issues including:
- Missing values
- Group bias
- Distribution drift

The goal is to validate the methods implemented in the inspector package.
```

```python
import pandas as pd
import sys
from pathlib import Path

ref = pd.read_csv("../data/reference/sample.csv")
cur = pd.read_csv("../data/raw/sample.csv")

ref.head(), cur.head()
project_root = Path("..").resolve()
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))
```

```python
cur.describe(include="all")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(cur.isnull(), cbar=False)
plt.title("Missing Value Pattern in Current Dataset")
plt.show()
```

```python
from inspector.bias import group_distribution

group_distribution(cur, "gender", "approved")
```

```python
from inspector.drift import detect_drift

detect_drift(ref, cur)
```

```python
## Observations and Limitations

- Small sample sizes limit statistical power
- Different drift metrics capture different aspects of change
- Missing values affect certain statistical tests

Future work includes scaling analysis to larger datasets via openML.
```
